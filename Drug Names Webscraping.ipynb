```python
# Extracting list of drug URLs
base_url = "https://www.medicines.org.uk"
ingredient_url = "/emc/ingredient/"
drug_urls = []
for letter in "ABCDEFGHIJKLMNOPQRSTUVWXYZ0-9":
    # Generate the URL for each letter of the alphabet
    url = base_url + "/emc/browse-ingredients/" + letter.lower()
    # Send an HTTP GET request to the URL and store the response
    response = requests.get(url)
    # Get the HTML content from the response
    html_content = response.text

    # Regular expression pattern to match the desired URLs
    pattern = r'<a class="emc-link" href="([^"]+)">'
    # Find all matches of the pattern within the HTML content
    matches = re.findall(pattern, html_content)
    # Filter the URLs to only include ingredient URLs
    ingredient_urls = [base_url + match for match in matches if ingredient_url in match]
    # Append the extracted URLs to the drug_urls list
    drug_urls.extend(ingredient_urls)
```

```python
#From the drug_urls let's extract the end point parameters to make the code a bit faster first
# Regular expression pattern to match the last part of the drug_urls for url IDs
pattern = r'/emc/ingredient/(\d+)'

# List to store the extracted url IDs
url_ids = []

# Iterate over the drug URLs
for url in drug_urls:
    # Find the url ID from the URL using the regular expression pattern
    match = re.search(pattern, url)
    if match:
        url_id = match.group(1)
        url_ids.append(url_id)

# Print the extracted ingredient IDs
for url_id in url_ids:
    print(url_id)
```

```python
#Importing concurrent features to speed up the code in looping over each url html content
import concurrent.futures
```

```python

base_url = 'https://www.medicines.org.uk/emc/ingredient/{}'
drug_names = set()  # Use a set to store unique drug names then convert to a list

# Define a function to fetch the drug name for a given ID
def fetch_drug_name(url_id):
    url = base_url.format(url_id)
    response = requests.get(url)
    html_content = response.text

    pattern = r'<a class="search-results-product-info-title-link emc-link" href="/emc/product/\d+/[a-z]+">\s*(.*?)\s*</a>'
    data = re.findall(pattern, html_content)
    drug_list = set([name.strip().upper() for name in data])  # Normalize and remove duplicates
    
    return drug_list

# Use ThreadPoolExecutor to execute the requests concurrently
with concurrent.futures.ThreadPoolExecutor() as executor:
    # Submit each ID to the executor
    futures = [executor.submit(fetch_drug_name, url_id) for url_id in url_ids]
    
    # Retrieve the results as they complete
    for future in concurrent.futures.as_completed(futures):
        drug_names.update(future.result())  # Use the 'update' method of set to add unique drug names

# Print the extracted drug names
for drug_name in drug_names:
    print(drug_name)
```

```python
#Extracting list of drug names
base_url ="https://www.medicines.org.uk/emc/browse-medicines/{}"
medicines=[]
for letter in "ABCDEFGHIJKLMNOPQRSTUVWXYZ0-9":
    # Generate the URL for each letter of the alphabet
    url = base_url.format(letter.lower())
    
    # Send an HTTP GET request to the URL and store the response
    response = requests.get(url)
    
    # Get the HTML content from the response
    html_content = response.text

    # Regular expression pattern to match the desired drug names
    pattern = r'<a class="search-results-product-info-title-link emc-link" href="/emc/product/\d+/smpc">\s*(.*?)\s*</a>'

    # Find all matches of the pattern within the HTML content
    data = re.findall(pattern, html_content)

    #  Normalize and remove duplicates
    unique_drug_names = list(set([name.strip().upper() for name in data]))
    
    #Append to the medicines list

    medicines.extend(unique_drug_names)

```

```python
#Print out the drug names
medicines
```

```python
#Combine both lists and remove duplicates
import numpy as np
drugs=np.unique(drug_names+medicines)

#Convert the combined lists to dataframe
# Create a DataFrame from the drug_list
drugs_data = pd.DataFrame(drugs, columns=['Name'])
drugs_data
```
